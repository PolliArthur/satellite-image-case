## Importação de libs


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

## Funções utilitárias

```python
def plot_side_by_side(image1, image2, title1='Image 1', title2='Image 2'):
    """
    Plots two images side by side using Matplotlib.

    Args:
        image1 (numpy.ndarray): The first image to be plotted.
        image2 (numpy.ndarray): The second image to be plotted.
        title1 (str): Title for the first image.
        title2 (str): Title for the second image.
    """
    plt.figure(figsize=(12, 5))  
    plt.subplot(1, 2, 1)
    plt.imshow(image1)
    plt.title(title1)
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(image2)
    plt.title(title2)
    plt.axis('off')

    plt.show()
```

## Remoção de ruídos

```python
#leitura da imagem com openCV
image = cv2.imread('../data/q1/img46_gray_noise.png')
# visualização usando matplotlib
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')  
plt.show()
```

**Filtro de Mediana:**
O filtro de mediana é um tipo de filtro passa-baixa que substitui cada pixel na imagem pela mediana dos valores dos pixels na vizinhança. Ele é eficaz na remoção de ruído impulsivo ou "salt-and-pepper".

### Testes de tamanho de  (3x3),(5x5) e (7x7)

```python
median_blur_imgs = {"Kernel_3":{'img':None,'k_size':3}
                    ,"Kernel_5":{'img':None,'k_size':5}
                    ,"Kernel_7":{'img':None,'k_size':7}}

for key, item in median_blur_imgs.items():      
    item['img']  = cv2.medianBlur(image, item['k_size'])    
    plot_side_by_side(image, item['img'],
                       'Gray Noise Image', 
                      f"Median Blur Image ({item['k_size']}x{item['k_size']})")
```

## Comparando melhor resultado com a img46_gray.png
Considerando a imagem obtida com kernel = 5x5 como melhor resultado, visto que foi possível eliminar os ruídos e não borrou tanto a imagem quanto a 7x7.

```python
img46 = cv2.imread('../data/q1/img46_gray_noise.png')
plot_side_by_side(img46, median_blur_imgs["kernel_5"], 'Image 46 gray', f"Median Blur Image (5x5)")
```
